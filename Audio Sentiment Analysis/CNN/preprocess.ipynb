```python
import os
import pandas as pd
import librosa
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
from sklearn.preprocessing import MinMaxScaler
```

```python
CREMA=os.listdir('CREMA/')
SAVEE=os.listdir('SAVEE/')
TESS=os.listdir('TESS/TESS Toronto emotional speech set data')
RAVEDESS=os.listdir('RAVEDESS/audio_speech_actors_01-24/')
```

```python
emotion= []
path= []
for audio_file in CREMA:
    if "ANG" in audio_file:
        emotion.append("Angry")
        path.append("CREMA/"+str(audio_file))
    if "HAP" in audio_file:
        emotion.append("Happy")
        path.append("CREMA/"+str(audio_file))
    if "SAD" in audio_file:
        emotion.append("Sad")
        path.append("CREMA/"+str(audio_file))
    if "FEA" in audio_file:
        emotion.append("Fear")
        path.append("CREMA/"+str(audio_file))
    if "DIS" in audio_file:
        emotion.append("Disgust")
        path.append("CREMA/"+str(audio_file))
    if "NEU" in audio_file:
        emotion.append("Neutral")
        path.append("CREMA/"+str(audio_file))
```

```python
ID = [(file.split("/"))[1][:4] for file in path]
#Phrase = [(file.split("_"))[1] for file in path]
Emotion = [(file.split("_"))[2] for file in path]
#Level = [(file.split("_"))[3][:2] for file in path]
```

```python
F_ID = ['1002','1003','1004','1006','1007','1008','1009','1010','1012','1013','1018','1020','1021','1024','1025','1028',
        '1029','1030','1037','1043','1046','1047','1049','1052','1053','1054','1055','1056','1058','1060','1061','1063',
        '1072','1073','1074','1075','1076','1078','1079','1082','1084','1089','1091']
Gender=[]

for i in range(0,len(ID)):
    if ID[i] in F_ID:
        Gender.append('Female')
    else:
        Gender.append('Male')
```

```python
CREMA_df = pd.DataFrame({
    'Emotion': emotion,
    'Path': path,
    'Gender':Gender
})
```

```python
emotion=[]
path=[]
Gender=[]
for audio in SAVEE:
    if audio[-8:-6]=='_a':
        emotion.append("Disgust")
        path.append("SAVEE/"+str(audio))
        Gender.append('Male')
    elif audio[-8:-6]=='_f':
        emotion.append('Fear')
        path.append("SAVEE/"+str(audio))
        Gender.append('Male')
    elif audio[-8:-6]=='_h':
        emotion.append('Happy')
        path.append("SAVEE/"+str(audio))
        Gender.append('Male')
    elif audio[-8:-6]=='_n':
        emotion.append('Neutral')
        path.append("SAVEE/"+str(audio))
        Gender.append('Male')
    elif audio[-8:-6]=='sa':
        emotion.append('Sad')
        path.append("SAVEE/"+str(audio))
        Gender.append('Male')
    elif audio[-8:-6]=='su':
        emotion.append('Surprising')
        path.append("SAVEE/"+str(audio))
        Gender.append('Male')
    
    
```

```python
SAVEE_df = pd.DataFrame({
    'Emotion': emotion,
    'Path': path,
    'Gender':Gender
})
```

```python
SAVEE_df
```

```python
emotion=[]
path=[]
Gender=[]
for audio in RAVEDESS:
    actordir=os.listdir('RAVEDESS/audio_speech_actors_01-24/' + str(audio))
    for actor in actordir:
        file=actor.split(".")[0].split("-")
        emotion.append(int(file[2]))
        genderval = int(file[6])
        Gender.append("Female" if genderval % 2 == 0 else "Male")
        path.append('RAVEDESS/audio_speech_actors_01-24/' + str(audio)+ "/" +str(actor))
```

```python
RAVEDESS_df = pd.DataFrame({
    'Emotion': emotion,
    'Path': path,
    'Gender':Gender
})
RAVEDESS_df= RAVEDESS_df.replace({1:'Neutral', 2:'Neutral', 3:'Happy', 4:'Sad', 5:'Angry', 6:'Fear', 7:'Disgust', 8:'Surprise'})
```

```python
emotion=[]
path=[]
Gender=[]
dir_emotion_map = {
    'OAF_angry': 'Female_Angry',
    'YAF_angry': 'Female_Angry',
    'OAF_disgust': 'Female_Disgust',
    'YAF_disgust': 'Female_Disgust',
    'OAF_fear': 'Female_Fear',
    'YAF_fear': 'Female_Fear',
    'OAF_happy': 'Female_Happy',
    'YAF_happy': 'Female_Happy',
    'OAF_neutral': 'Female_Neutral',
    'YAF_neutral': 'Female_Neutral',
    'OAF_pleasant_surprise': 'Female_Surprise',
    'YAF_pleasant_surprise': 'Female_Surprise',
    'OAF_sad': 'Female_Sad',
    'YAF_sad': 'Female_Sad'
}

for audio in TESS:
    dir_path = os.listdir('TESS/TESS Toronto emotional speech set data/' + str(audio))
    for file in dir_path:
        emotionval = dir_emotion_map.get(audio, 'Unknown')
        emotion.append(emotionval)
        path.append('TESS/TESS Toronto emotional speech set data/' + str(audio) +"/" + str(file)) 
```

```python
TESS_df = pd.DataFrame({
    'Emotion': emotion,
    'Path': path
})
```

```python
TESS_df[["Gender","Emotion"]]= TESS_df["Emotion"].str.split("_", n=1, expand=True)
```

```python
TESS_df = TESS_df.dropna(subset=["Emotion"])
```
